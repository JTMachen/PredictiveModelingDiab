```python
data = pd.read_csv("diabetes_dataset__2019.csv")
data.head()
```

```python
## This cell cleans up the dataframe, since there are several instances of "no" vs. " no" that mean the same thing, but are classified as different things
```

```python
type_list = []
for yn in data.Pdiabetes:
    if yn == "0":
        yn = "no"
    type_list.append(yn)
data['Pdiabetes'] = type_list
type_list = []
for yn in data.RegularMedicine:
    if yn == "o":
        yn = "no"
    type_list.append(yn)
data['RegularMedicine'] = type_list
type_list = []
for yn in data.Diabetic:
    if yn == " no":
        yn = "no"
    type_list.append(yn)
data['Diabetic'] = type_list
type_list = []
for yn in data.BPLevel:
    if yn == "normal ":
        yn = "normal"
    elif yn == "High":
        yn = "high"
    elif yn == "Low":
        yn = "low"
    type_list.append(yn)
data['BPLevel'] = type_list
data.drop(columns = "BMI",inplace=True)
data['Pregancies'].fillna(0,inplace=True)
data.dropna(inplace=True)
data.info()
```

As the above cell shows, there are now 950 non-null columns, given that the two null rows in both the Pdiabetes and Diabetic columns have been removed.

```python
#Logisitc Regression no scaling
X = data.drop(['Diabetic'], axis=1)
y = data['Diabetic']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

lr = LogisticRegression()
lr.fit(X_train,y_train)

predictions = lr.predict(X_test)

log_reg = confusion_matrix(y_test, predictions)
print('\nClassification report\n', classification_report(y_test, predictions))
lr_score = round(r2_score(y_test,predictions),2)
display_markdown(f"According to the results, the regression score for the logistic regression model is {lr_score}, while the weighted average accuracy is {classification_report(y_test, predictions).split('weighted avg')[1][9:11]}%.",raw=True)
```

```python
#Logisitc Regression Robust Scaler
X = data.drop(['Diabetic'], axis=1)
y = data['Diabetic']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

scaler = RobustScaler()
X_train = scaler.fit_transform(X_train )
X_test = scaler.transform(X_test)

lr = LogisticRegression()
lr.fit(X_train,y_train)

predictions = lr.predict(X_test)

lr_stand = confusion_matrix(y_test, predictions)
print('\nClassification report\n', classification_report(y_test, predictions))
lr_stand_score = round(r2_score(y_test,predictions),2)
display_markdown(f"According to the results, the regression score for the robust scaled logistic regression model is {lr_stand_score}, while the weighted average accuracy is {classification_report(y_test, predictions).split('weighted avg')[1][9:11]}%.",raw=True)
```

```python
## Run the classification (since this is a classification problem) using both robust and standard scaler, choosing the best and using that particular
## scaler in the future algorithms
```

```python
X = data[['Age', 'Gender', 'Family_Diabetes', 'highBP', 'PhysicallyActive','Smoking', 'Alcohol', 'Sleep', 'SoundSleep', 'RegularMedicine','JunkFood', 'Stress', 'BPLevel', 'Pregancies', 'Pdiabetes','UriationFreq']]
y = data['Diabetic']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
stand = StandardScaler()
X_train = stand.fit_transform(X_train )
X_test = stand.transform(X_test)
scores = []
matricies_scaled = []
classification_list = []
for x in range(50):
    mlp = MLPClassifier(max_iter=500, activation='relu',hidden_layer_sizes=(128,64,32))
    mlp.fit(X_train,y_train)
    predictions = mlp.predict(X_test)
    scores.append(r2_score(y_test,predictions))
    cm_scale = confusion_matrix(y_test,predictions)
    classif = classification_report(y_test,predictions)
    matricies_scaled.append(cm_scale)
    classification_list.append(classif)
standard_ave = round(sum(scores)/len(scores),2)
standard_max = round(max(scores),2)
cm_standard = matricies_scaled[pd.Series(scores).idxmax()]
print("                   STANDARD SCALER")
print("         Best Standard Classification Report")
display(classification_list,scores)
print("\nMaximum Regression Score:",round(max(scores),2))
print("Average Regression Score:",round(sum(scores)/len(scores),2))
print("Minimum Regression Score:",round(min(scores),2))
scores = cross_val_score(mlp, X_train, y_train, cv = 10, scoring='accuracy')
print('Average cross-validation score: {:.2f}'.format(scores.mean()))
print("\n")
display_markdown(f"Based on 50 iterations of the classification algorithm, the average regression score for the standard scaler is {standard_ave}, while the best weighted average accuracy is {classification_list[pd.Series(scores).idxmax()].split('weighted avg')[1][9:11]}%.",raw=True)
```
